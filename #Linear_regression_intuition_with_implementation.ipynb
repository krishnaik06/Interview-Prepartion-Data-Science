```python
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=load_boston()
df
```

```python
dataset=pd.DataFrame(df.data) # for viewing features of dataset
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
## independent features and dependent features(we have divided data)
X=dataset
y=df.target
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size= 0.30,random_state=42)
```

```python
X_train
```

DO we have to standarize or normalize the data??
>>there is a formation of gradient decent in linear regression so we need to standarize the data . for which we use standardScalar


```python
# standarizing the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
#############  when we apply crossvalidation in linear regression we apply mean while in classification we apply modeeeeeeeeeee
```

>> in crossvalidation we divide model into multiple train test and each crossvalidation will have different output of accuracy.Result would be taken by mean of accuracies.

```python
from sklearn.linear_model import LinearRegression
#cross val
from sklearn.model_selection import cross_val_score 
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)#scoring is acuuracy's parameter
#regression case lose func is mean sq error & mean abs error
```

```python
np.mean(mse) # it ashould be less cuz it is difference of predicted and truth value
```

```python
##prediction
reg_predic=regression.predict(X_test)
```

```python
reg_predic
```

```python
### for checking whether the prediction is correct or not then we compare specifically from truth value ie. y_test
```

```python
import seaborn as sns
sns.displot(reg_predic-y_test,kind='kde')
#hence d differnce btwn prediction n test is good , its variance is less -10 to 10
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_predic,y_test)
score
```

>> DIFFERNCE BETWEEN TRANSFORM AND FIT_TRANSFORM
FIT:- transformer learns something about the data
TRANSFORM:- it uses what it learned to do the data transformation

* CountVectorizer
. fit learns the vocabulary
. transform creates a document-term matrix using the vocabulary

* SimpleImputer
. fit learns the value to impute
. transform fills in missing entries using the imputation value

* StandardScaler
. fit learns the mean and scale of each feature
. transform standardizes the features using the mean and scale
